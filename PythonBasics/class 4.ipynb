```python
# create a dictionary from lists
keys = ['person1','person2','person3']
values = ['Python','c++','js']
dict1 = dict(zip(keys,values)) #zip() keyword maps and combines both the lists for converting to a dictinonary
dict1
```

```python
dict1['person1']
```

```python
dict1['person3']
```

```python
#add key:value pairs
dict1['person4'] = 'Java'
```

```python
dict1
```

```python
#remove key:value pairs
del dict1['person2']
```
