```python
# reverse a list (two methods)
c = [3,6,2,1,7,9,8,4,6,5,2]
#1st method
c.reverse()
c
```

```python
#2nd method to reverse
c[::-1]   #select all the values from start to stop and print them in a reverse order (in steps of -1)
```

```python
c
```

```python
c[2::-2]  #reverse the list and print alternate numbers
```

```python
#operations on a list
append = add element at the end of list
insert = add element wrt index
remove = delete the first occurence of the element
pop    = delete element wrt index
```

```python
fruits = ['orange','mango','apple','banana','grapes']
fruits
```

```python
fruits.append('apple') #adds the element towards the end of the list
fruits
```

```python
fruits[0]  # indexing
```

```python
fruits.index('mango')
```

```python
# 'orange', 'mango', 'apple', 'banana', 'grapes', 'apple'
  #  0        1         2        3        4          5        -     6 elements (0-5 index)
  # -6       -5        -4       -3       -2         -1        -     negative indexing
```

```python
fruits[1:4]   #    1 - inclusive : 4 - exclusive  (range of values)
```
