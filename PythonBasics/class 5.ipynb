```python
# while loop 
# syntax 

# initialization 
# while condition:
#   statements/expression
#   increment/decrement 
```

```python
i = 0          #initialization       (0 - start, 10 - stop)
while i<10:    #while condition      #the control enters the while loop only if the condition is True    
  # print(i)     #statement
  print(i,end = " ")     # by default python has got the new line in the print statement (end removes that new line and we are replacing it with a space)
  i = i + 1    #increment in steps of 1
  # i += 1
  # i++ #this gives an error 
```

```python
i = 0       
while i<=20:         
  print(i)     
  i = i + 2   
```

```python
# 0 to 20 - odd numbers using while loop 
# post increment (first execute the statements,then increment)             #   (a++) - post increment                (++a) - pre increment   

i = 1
while i <= 20:
  print(i)
  i = i + 2   
```

```python
# pre increment (first increment,then execute the statements) 

i = 1
while i <= 20:
  i = i + 2
  # ++i (doesn't work)
  print(i)
```

```python
# endless loop / forever loop
while True:
  print("HELLO WORLD!")
```

```python
# post decrement 
# (start - number should be greater, stop - number should be smaller)

i = 10
while i >= 0:   
  print(i)
  i = i - 1
```

```python
# pre decrement 

i = 10
while i >= 0:  
  i = i - 1 
  print(i)
```

```python
# tables using a while loop 
n = int(input("Enter the number : "))
i = 1
while i <=10:
  print(f"{n} x {i} = {n*i}")
  i = i + 1
```
