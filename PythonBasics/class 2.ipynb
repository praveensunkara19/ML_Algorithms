```python
a = int(input("Enter a number :"))
if a%2 == 0:
  print(f"{a} is Even!")
else:
  print(f"{a} is Odd!")
```

```python
# using if else check age eligibility of a person to vote
a = int(input("Enter your age :"))
if a >=18 :
  print("Eligible to vote!")
  print(f"Eligible to vote since last {a-18} years!")
else:
  print("Not eligible to vote!")
  print(f"Eligible to vote after {18-a} years!")
```

```python
# Elseif conditions in python uses the keyword elif
a = input("Enter the season :")
if a == "summer":
  print("Summer season!")
elif a == "winter":
  print("Winter season!")
elif a == "autumn":
  print("Autumn season!")
elif a == "spring":
  print("Spring season!")
else:
  print("Invalid input!")
```
