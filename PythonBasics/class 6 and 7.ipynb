```python
# star patterns using while 

# *
# * *
# * * *
# * * * *
# * * * * *
```

```python
n = int(input("Enter the number of rows :"))
i = 0 
while i<n:
  print(i*"* ")
  i = i + 1
```

```python
import time 
n = int(input("Enter the number of rows :"))
i = 0 
while i<n:
  print(i*"* ")
  time.sleep(1)  #adds a delay of 1 second after every iteration
  i = i + 1
```

```python
n = 0
i = 5 
while i>n:
  print(i*"* ")
  i = i - 1
```

```python
#iterate over the dictionary keys
for i in {1:'aa',2:'bb',3:'cc',4:'dd'}:
  print(i)
```

```python
#iterate over the dictionary values
data = {1:'aa',2:'bb',3:'cc',4:'dd'}

for i in data.values():
  print(i)
```

```python
#iterate over the dictionary items
data = {1:'aa',2:'bb',3:'cc',4:'dd'}

for i in data.items():
  print(i)
```

```python
# tables using a for loop 
n = int(input("Enter a number : "))
for i in range(1,11,1):
  print(f"{n} x {i} = {n*i}")
```

```python
def add(x,y):
  sum = x+y
  return sum 

add(2,3)
```

```python
def add_sub(x,y):
  sum = x+y
  difference = x-y 
  return sum,difference

add_sub(7,4)
```
