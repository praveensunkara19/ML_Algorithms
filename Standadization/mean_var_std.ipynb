```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = [i for i in range(10)]
y = np.random.randint(0,10, 10)
print(x,y)
```

```python
plt.scatter(x,y)
```

```python
m = np.mean(y)
var = np.var(y)
std = np.std(y)
print(m)
print(var)
print(std)
```

Varience

```python
def varience(y):
    return np.sum((y-m)**2)/10

v = varience(y)
print(v)
```

Standard Deviation

```python
def standardDeviation(y):
    return math.sqrt(np.sum((y-m)**2)/10)

s = standardDeviation(y)
print(s)
```

Calculating Mean, StandardDeviation

```python
y = [0, 0, 3, 3, 7, 3, 8, 1, 0, 7]
x = np.arange(len(y))  
mean = np.mean(y)
sd = np.std(y)         
var = np.var(y)

plt.scatter(x, y, color='blue', label='Data Points')
plt.axhline(mean, color='red', linestyle='--', label=f'Mean ({mean})')

# Visualizing SD as a shaded "corridor"
plt.axhspan(mean - sd, mean + sd, color='gray', alpha=0.2, label='1 SD Band')

plt.title('Scatter Plot with Mean and SD')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(100)  
y = np.random.randint(0,100,100)
mean = np.mean(y)
sd = np.std(y)         

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', zorder=5, label='Data Points')
plt.axhline(mean, color='red', linestyle='--', label=f'Mean ({mean:.2f})')

# 1. Visualize SD (The 'Corridor') -> typical points that falls under the varience in most of datasets 68% in the range of varince
plt.axhspan(mean - sd, mean + sd, color='gray', alpha=0.2, label=f'1 SD Band (±{sd:.2f})')

# 2. Visualize Variance (The 'Squares')
for i in range(len(y)):
    diff = y[i] - mean
    # Draw a square where the side length is the distance to the mean
    rect = plt.Rectangle((x[i], mean), diff, diff, color='orange', alpha=0.2, lw=1)
    plt.gca().add_patch(rect)

plt.title('Scatter Plot: Mean, SD (Band), and Variance (Squares)')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

Histogram Representation

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.normal(100, 10, 1000)
mean = np.mean(data)
sd = np.std(data)

# Histogram
plt.hist(data, bins=30, alpha=0.5, color='blue', edgecolor='black')

# Mean line
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label='Mean')

# SD lines
plt.axvline(mean - sd, color='green', linestyle='dotted', linewidth=2, label='$\pm1$ SD')
plt.axvline(mean + sd, color='green', linestyle='dotted', linewidth=2)
plt.legend()
plt.title('Visualization of Mean and SD')
plt.show()
```
