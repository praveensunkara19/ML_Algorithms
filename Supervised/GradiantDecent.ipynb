Using mse error, deriving the values of m,c from y = m * x + c . the Gradient Decent finds the best fit line 
by minimising the m,c with the slope of the tangent along the graph curve of the [mse(cost funciton),m,c] in order to find the minimun radient value point(global minima) with iterations.
At this minima the values of m,c is the variables used to find the best fit line. 

for every iteration the new values of m,c is 

m = m - lr * f`(m)
c = c - lr * f`(c)



```python
x = [1,2,3,4,5]

y = [5,7,9,11,13]

# need to predict new values of y with the x input using gradient decent 
```

```python
import numpy as np 

# we use numpy arrays coz faster than list
x = np.array(x)
y = np.array(y)
```

```python
def gradient_decent(x,y):
    m_curr = c_curr = 0
    iterations = 1000
    n = len(x) #~len(y) assuming len(x) = len(y)
    lr = 0.01
    for i in range(iterations):
        y_pred = m_curr * x + c_curr
        md = -2/n * sum(x*(y-y_pred))   #d(mse)/dm
        cd = -2/n * sum(y-y_pred)       #d(mse)/dc
        m_curr = m_curr - lr * md
        c_curr = c_curr - lr * cd
        print("m: {} c: {} iteration {}".format(m_curr,c_curr,i))

gradient_decent(x,y)
```

```python
def gradient_decent(x,y):
    m_curr = c_curr = 0
    iterations = 1000
    n = len(x) #~len(y) assuming len(x) = len(y)
    lr = 0.0001
    for i in range(iterations):
        y_pred = m_curr * x + c_curr
        cost = 1/n * sum(val**2 for val in (y - y_pred) )  #mse
        md = -2/n * sum(x*(y-y_pred))   #d(mse)/dm
        cd = -2/n * sum(y-y_pred)       #d(mse)/dc
        m_curr = m_curr - lr * md
        c_curr = c_curr - lr * cd
        print("m: {} c: {} iteration {} cost: {}".format(m_curr,c_curr,i,cost))

gradient_decent(x,y)
```
