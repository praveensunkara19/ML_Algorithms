```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
dir(digits)
```

```python
digits.data[0]
```

```python
import matplotlib.pyplot as plt

for i in range(5):
    plt.gray()
    plt.imshow(digits.images[i])
    plt.show()
```

```python
digits.target
```

```python
x = digits.data
y = digits.target
```

```python
print(len(x))
print(len(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
import pandas as pd

pd.DataFrame(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

Random Forest Model has more accuracy than Decsion Tree because, multiple decision trees are formed form the dataset in batches

## Fine tuning by changing the paramerter values

```python
model1 = RandomForestClassifier(n_estimators=500,criterion='entropy')
```

```python
model1.fit(x_train,y_train)
```

```python
model1.score(x_train,y_train)
```

```python
y_pred1 = model1.predict(x_test)
```

```python
accuracy_score(y_test,y_pred1)
```
