```python
import pandas as pd

df = pd.read_csv("C:\\Users\\prave\\OneDrive\\Desktop\\Git\\MachineLearning\\Datasets\\salaries.csv")
df
```

```python
x = df.iloc[:,0:3]
y = df.iloc[:,-1]
```

```python

print(x.ndim)
print(y.ndim)
```

```python
#In machine learning model learn from numbers only. so, we need to convert the words to numbers

from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
x["company_n"] = le_company.fit_transform(x['company'])
x["job_n"] = le_job.fit_transform(x['job'])
x["degree_n"] = le_degree.fit_transform(x['degree'])
```

```python
x
```

```python
new_x = x.iloc[:,3:6]
new_x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(new_x,y,test_size=15)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Dtree = DecisionTreeClassifier()
```

```python
Dtree.fit(x_train,y_train)
```

```python
Dtree.score(x_train,y_train) #model score
```

```python
y_pred = Dtree.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
