```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Matlab - similar to Matplotlib (Matplotlib is a data visulaization library)
```

```python
x = np.arange(0,30,0.1)
y = np.sin(x)

# plt.plot(x,y)

# to fill the sinewave with any color
fig,ax = plt.subplots()   # create subplots (plot inside a image space)
ax.plot(x,y)
ax.fill_between(x,y,color='blue')

plt.show()
```

```python
#LINE GRAPH

a = [1,2,3,4,5,6]
b = [5,4,5,7,3,2]
c = [1,4,8,5,2,3]

plt.plot(a,b,color = 'red',label = "Line 1",marker = "<",linewidth = 5,markerfacecolor = "black",markeredgecolor="black",markersize = 10)
plt.plot(a,c,color = 'blue',label = "Line 2",linestyle = "dashdot")  #linestyle - solid,dotted,dashed,dashdot
plt.title("LINE GRAPH")
plt.xlabel("X - AXIS")
plt.ylabel("Y - AXIS")
plt.legend()
plt.show()
```

```python
#BAR GRAPH
x = ["A","B","C","D","E"]
height = [150,135,174,120,197]
# plt.bar(x,height,color = "red")
plt.bar(x,height,color = ["red","blue","green","violet","brown"])
```

```python
#BAR GRAPH
x = ["A","B","C","D","E"]
height = [150,135,174,120,197]
plt.barh(x,height,color = "red")
```

```python
#Scatter plot
a = [1,2,3,4,5,6,7,8]
b = [5,4,7,3,2,9,1,3]
plt.scatter(a,b,c = ['r','g','b','c','m','y','k','brown'], s = 150, marker = "^",alpha = 0.5) #alpha is level of transparency (0-1)
```

```python
# random values for scatter plot
# np.random.seed(67564) #keeps values fixed at some random integers

a= np.random.randint(0,50,100) #100 random values from 0 to 50
b = np.random.randint(0,50,100) #100 random values from 0 to 50
plt.scatter(a,b,marker = "*",s = 150,c = a,cmap = "jet")
plt.show()

# https://matplotlib.org/stable/tutorials/colors/colormaps.html
```
